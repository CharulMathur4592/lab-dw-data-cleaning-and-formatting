## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*

```python
# Your code here

import pandas as pd
import re

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")


df.columns = (df.columns
    .str.strip()
    .str.lower()
    .str.replace(r"\s", "_", regex = True)
)

df = df.rename(columns={
    "st": "state",
    "amt": "amount",
})

print(df.columns.tolist())
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here
def clean_colname(col):
    col = col.strip()
    col = col.lower()
    col = re.sub(r"\s+", "_", col)
    col = re.sub(r"[^\w_]", "", col)         
    col = re.sub(r"_+", "_", col)
    return col

df.columns = [clean_colname(c) for c in df.columns]

print("Columns after normalization:")
print(df.columns.tolist())
print("\n---\n")

column_names = ["gender", "state", "education", "customer_lifetime_value", "vehicle_class"]
print("Column Names Before Cleaning: ")

for c in column_names:
    if c in df.columns:
        print(f"\n{c}: ")
        print(df[c].astype(str).value_counts(dropna = False).head(30))
    else:
        print(print(f"\n{c}: (column not present)"))
print("\n---\n")

# Cleaning Gender Column
gender_map = {
    "female": "F", "f": "F", "femal": "F", "fem": "F", "fe": "F", "woman": "F",
    "female ": "F", "female\t": "F",
    "male": "M", "m": "M", "maile": "M", "male ": "M", "man": "M"
}

def clean_gender_col(x):
    if pd.isna(x):
        return x

    s = str(x).strip().lower()
    if s in gender_map:
        return gender_map[s]

    if s in ("m", "f"):
        return s.upper()

    if s and s[0] in ("m", "f"):
        return [0].upper()
    return x

if "gender" in df.columns:
    df["gender"] = df["gender"].apply(clean_gender_col)

# Cleaning State Column
state_map = {
    "az": "Arizona",
    "ca": "California",
    "wa": "Washington",
    "ny": "New York",
    "cali": "California"
}

def clean_state_col(x):
    if pd.isna(x):
        return x
        
    s = str(x).strip().lower()

    s_clean = re.sub(r"[^\w]", "", s)
    return state_map.get(s_clean, x)

if "state" in df.columns:
    df["state"] = df["state"].apply(clean_state_col)

# Cleaning Education Column
education_map = {
    "bachelor's": "Bachelor",
    "masters": "Master",
    "phd": "PhD",
    "highschool": "High School",
    "high school": "High School",
    "associate": "Associate",
    "bachelors": "Bachelor"
}

def clean_education_col(x):
    if pd.isna(x):
        return x
    
    s = str(x).strip()
    
    key = re.sub(r"[^\w]", "", s.lower())
    return education_map.get(key, s)

if "education" in df.columns:
    df["education"] = df["education"].apply(clean_education_col)

# Cleaning Customer lifetime value Column
def clean_clv(x):
    if pd.isna(x):
        return x
        
    s = str(x).strip()
    s = s.replace("%", "").replace(",", "")

    try:
        return float(s)
    except:
        return x

if "customer_lifetime_value" in df.columns:
    df["customer_lifetime_value"] = df["customer_lifetime_value"].apply(clean_clv)

# Cleaning Vehicle column
vehicle_map = {
    "sports car": "Luxury",
    "luxury suv": "Luxury",
    "luxury car": "Luxury",
    "luxury": "Luxury",
    "sports": "Sports",
}

def clean_vehicle_column(col):
    if pd.isna(col):
        return col
        
    s = str(col).strip().lower()
    s_nopunct = re.sub(r"[^\w\s]", "", s)

    if s_nopunct in vehicle_map:
        return vehicle_map[s_nopunct]

    if "luxury" in s_nopunct:
        return "Luxury"
    if "sport" in s_nopunct:
        return "Sports"
    return s.title()

if "vehicle_class" in df.columns:
    df["vehicle_class"] = df["vehicle_class"].apply(clean_vehicle_column)

print("Column names AFTER cleaning: ")
for t in column_names:
    if t in df.columns:
        print(f"\n{t}:")
        print(df[t].astype(str).value_counts(dropna=False).head(50))
    else:
        print(f"\n{t}: (column not present)")
```

## Exercise 3: Formatting data types

The data types of many columns in the dataset appear to be incorrect. This could impact the accuracy of our analysis. To ensure accurate analysis, we need to correct the data types of these columns. Please update the data types of the columns as appropriate.

It is important to note that this exercise does not involve handling null values (NaN or None).

*Hint*:
- *Customer lifetime value should be numeric*
- *Number of open complaints has an incorrect format. Look at the different values it takes with `unique()` and take the middle value. As an example, 1/5/00 should be 5. Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. Finally, since it should be numeric, cast the column to be in its proper type.*

```python
# Your code here

print("Initial dtypes:")
print(df.dtypes)
print("\nColumns:")
print(df.columns.tolist())

# identify likely column names (adjust if your file uses slightly different names)
clv_col = "customer_lifetime_value"
complaints_col = None

# try guesses for complaints column name
for candidate in ["number_of_open_complaints", "number_of_open_complaint", "open_complaints", "num_open_complaints"]:
    if candidate in df.columns:
        complaints_col = candidate
        break

print("\nFound columns:")
print("customer_lifetime_value ->", clv_col in df.columns)
print("number_of_open_complaints ->", complaints_col)

# Show unique samples BEFORE cleaning
if clv_col in df.columns:
    print("\nSample unique values for customer_lifetime_value (BEFORE):")
    print(df[clv_col].astype(str).value_counts(dropna=False).head(40))

if complaints_col:
    print("\nSample unique values for", complaints_col, "(BEFORE):")
    print(df[complaints_col].astype(str).value_counts(dropna=False).head(60))
else:
    print("\nCould not find complaints column automatically. Check column names and update the script's candidate list.")

# 2) Convert Customer Lifetime Value -> numeric
def clean_clv_to_numeric(x):
    if pd.isna(x):
        return x
    s = str(x).strip()
    # remove percent / currency characters and thousands separators
    s = s.replace("%", "").replace("$", "").replace(",", "")
    # take the first numeric substring (handles trailing text)
    m = re.search(r"-?\d+(\.\d+)?", s)
    if m:
        return float(m.group(0))
    # fallback attempt
    try:
        return float(s)
    except:
        return pd.NA

if clv_col in df.columns:
    df["customer_lifetime_value_cleaned"] = df[clv_col].apply(clean_clv_to_numeric)
    # cast to numeric (float). If you want integers, cast after inspecting.
    df["customer_lifetime_value_cleaned"] = pd.to_numeric(df["customer_lifetime_value_cleaned"], errors="coerce")

# 3) Parse Number of Open Complaints: take the middle part of delimited values
def extract_middle_numeric(val):
    if pd.isna(val):
        return val
        
    s = str(val).strip()
    
    # if already a plain integer-looking string, return it
    if re.fullmatch(r"\d+", s):
        return int(s)
    
    # split on common separators (/, -, :, whitespace)
    parts = re.split(r"[\/\-\:\s]+", s)
    parts = [p for p in parts if p != ""]
    
    if not parts:
        return pd.NA
    
    mid_idx = len(parts) // 2  # e.g. for 3 parts, idx=1 -> middle
    
    candidate = parts[mid_idx]
    
    m = re.search(r"\d+", candidate)
    
    if m:
        return int(m.group(0))
    
    # fallback: scan parts for any numeric and return first
    for p in parts:
        m = re.search(r"\d+", p)
        if m:
            return int(m.group(0))
    return pd.NA

if complaints_col:
    df["number_of_open_complaints_cleaned"] = df[complaints_col].apply(extract_middle_numeric)
    
    # cast to pandas nullable integer if possible
    try:
        df["number_of_open_complaints_cleaned"] = df["number_of_open_complaints_cleaned"].astype("Int64")
    except Exception:
        # fallback to numeric with coercion
        df["number_of_open_complaints_cleaned"] = pd.to_numeric(df["number_of_open_complaints_cleaned"], errors="coerce")

# 4) Show AFTER summaries
print("\n--- AFTER cleaning samples ---")

if clv_col in df.columns:
    print("\ncustomer_lifetime_value (AFTER) value counts (top 40):")
    print(df["customer_lifetime_value_cleaned"].value_counts(dropna=False).head(40))

if complaints_col:
    print("\nnumber_of_open_complaints (AFTER) value counts (top 40):")
    print(df["number_of_open_complaints_cleaned"].value_counts(dropna=False).head(40))

# 5) Show dtypes for cleaned columns
print("\nDtypes of cleaned columns:")
for c in ["customer_lifetime_value_cleaned", "number_of_open_complaints_cleaned"]:
    if c in df.columns:
        print(c, "->", df[c].dtype)
```

## Exercise 4: Dealing with Null values

Identify any columns with null or missing values. Identify how many null values each column has. You can use the `isnull()` function in pandas to find columns with null values.

Decide on a strategy for handling the null values. There are several options, including:

- Drop the rows or columns with null values
- Fill the null values with a specific value (such as the column mean or median for numerical variables, and mode for categorical variables)
- Fill the null values with the previous or next value in the column
- Fill the null values based on a more complex algorithm or model (note: we haven't covered this yet)

Implement your chosen strategy to handle the null values. You can use the `fillna()` function in pandas to fill null values or `dropna()` function to drop null values.

Verify that your strategy has successfully handled the null values. You can use the `isnull()` function again to check if there are still null values in the dataset.

Remember to document your process and explain your reasoning for choosing a particular strategy for handling null values.

After formatting data types, as a last step, convert all the numeric variables to integers.

```python
# Your code here

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(r"\s+", "_", regex=True)
      .str.replace(r"[^\w_]", "", regex=True)
)

missing_counts = df.isnull().sum()
print("Null values per column: ", missing_counts)

df = df.dropna()

num_cols = df.select_dtypes(include = ["float", "int", "int64"]).columns
cat_cols = df.select_dtypes(include = ["object"]).columns

print("\nNumeric columns:", num_cols)
print("Categorical columns:", cat_cols)

for cols in num_cols:
    if df[cols].isnull().any():
        df[cols] = df[cols].fillna(df[cols].median())

for cols in cat_cols:
    if df[cols].isnull().any():
        mode = df[cols].mode(dropna = True)
        if len(mode) > 0:
            df[cols] = df[cols].fillna(mode[0])

print("\nAfter imputation, missing values per column:")
print(df.isnull().sum())

for cols in num_cols:
    df[cols] = df[cols].astype(int)
```
